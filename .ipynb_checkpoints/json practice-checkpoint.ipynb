```python
import requests
import bs4
import json
from urllib.request import urlopen
from pprint import pprint
```

```python
url = 'https://jsonplaceholder.typicode.com/todos'
```

```python
u = urlopen(url)
resp = json.loads(u.read().decode('utf-8'))
pprint(resp)
```
